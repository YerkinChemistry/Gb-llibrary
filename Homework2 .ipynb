В каждом графике массив x используется для координат по горизонтали.В левом верхнем графике для координат по вертикали используйте y1,в правом верхнем - y2, в левом нижнем - y3, в правом нижнем - y4.Дайте название графикам: 'График y1', 'График y2' и т.д.

```python
x = np.linspace(-3, 3, 51)
y1 = x**2
y2 = 2*x + 0.5
y3 = -3*x - 1.5
y4 = np.sin(x)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from pylab import rcParams
plt.style.use('fivethirtyeight')
```

```python
data = pd.read_csv("creditcard.csv", sep=',')
data
```

```python
#число мошенничеств и нормальных транзакции
fraudcounts = data.loc[:, 'Class'].value_counts()
```

```python
#Диаграмма
fraudcounts.plot(kind='bar')
plt.show()
```

```python
#Диаграмма с логарифмическим масштабом
fraudcounts.plot(kind='bar', logy=True)
plt.show()
```

```python
#Построение Гистаграммы для V1
fraud_hist = data.loc[data['Class'] == 1, 'V1']
norm_hist = data.loc[data['Class'] == 0, 'V1']
plt.hist(fraud_hist, bins=20, alpha=0.5, density=True, label='Class 1', color='red')
plt.hist(norm_hist, bins=20, alpha=0.5, density=True, label='Class 0', color='grey')
plt.xlabel("V1")
plt.legend()
plt.show()
```
